```python
from math import *
import random
from matplotlib import pyplot as plt
import numpy as np
from scipy.misc import derivative
```

```python
#task one
def func1(x):
    return tan(x)
def func2(x, n):
    return n*pi + atan(x)
def func3(x):
    return tan(x) - x
def func4(x):
    return x
```

```python
xmin = -10
xmax = 10
xlab = np.arange(xmin, xmax, 0.01)
ylab1 = np.array([func1(x) for x in xlab])
ylab2 = np.array([func4(x) for x in xlab])
plt.plot(xlab, ylab1)
plt.ylim(-10,10)
plt.plot(xlab, ylab2)                                    
plt.show()         
```

```python
#Bisection method
def Bis_method(a,b,delta,f):
    iterations = 0
    if f(a) == 0:
        return a,a, iterations
    if f(b) == 0:
        return a,b, iterations
    while(abs(a-b)> delta) and (iterations < 2000):
        if f(a) == 0:
            return a,a, iterations
        if f(b) == 0:
            return a,b, iterations
        c = (a+b)/2
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
        iterations += 1
    return a, (a + b)/2, iterations
```

```python
def Simple_iterations_method(a,b,delta,f,fi):
    iterations = 0
    if f(a) == 0:
        return a,a, iterations
    if f(b) == 0:
        return a,b, iterations
    x0 = random.random()*(b-a) + a
    x_prev = x0
    n = int(b/pi)
    x_next = func2(x0, n)
    iterations += 1
    while(abs(x_prev - x_next) > delta) and (iterations < 2000):
        x_ = x_next
        x_next = func2(x_next, n)
        x_prev = x_
        iterations += 1
    return x0, x_next, iterations
```

```python
def Newton_method(a,b,delta,f,fi):
    iterations = 0
    if f(a) == 0:
        return a,a, iterations
    if f(b) == 0:
        return a,b, iterations
    x0 = random.random()*(b-a) + a
    print(x0)
    x_prev = x0
    x_next = x0 - f(x0)/derivative(f, x0, dx = 1e-10)
    iterations += 1
    while(abs(x_prev - x_next) > delta) and (iterations < 2000):
        x_ = x_next
        x_next = x_next - f(x_next)/derivative(f, x_next, dx = 1e-6)
        x_prev = x_
        iterations += 1
    return x0, x_next, iterations
```

```python
def find_root(a, b, delta, f, fi):
    x0, a1, iters = Bis_method(a,b,delta,f)
    print(x0, a1, iters)
    x0, a1, iters = Simple_iterations_method(a,b,delta,f,fi)
    print(x0, a1, iters)
    x0, a1, iters = Newton_method(a,b,delta,f,fi)
    print(x0, a1, iters)
```
